```
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten

#宣告模型
model = Sequential() #宣告堆疊的模型
model.add(Flatten()) #二維輸入 轉為一維
model.add(Dense(512, activation='relu')) # 512 Neuron
model.add(Dropout(0.2)) #刪除20%輸出
model.add(Dense(10, activation='softmax')) #輸出0~9共10個可能性 SoftMax讓可能性總和為1
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) #編譯模型

import tensorflow as tf
mnist = tf.keras.datasets.mnist #讀入Mnist手寫數字資料庫
(x_train, y_train),(x_test, y_test) = mnist.load_data() #_traing是訓練資料6萬筆, _test是測試資料1萬筆
x_train, x_test = x_train / 255.0, x_test / 255.0 #正規化 

model.fit(x_train, y_train, epochs=5) #把訓練資料丟進訓練
model.evaluate(x_test, y_test) #評估測試資料的正確
```
